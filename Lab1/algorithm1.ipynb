```python
import matplotlib
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

import time
import matplotlib.pyplot as plt

# Реализация сортировки пузырьком
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Функция для замера времени выполнения
def measure_time(n):
    # Генерация случайного массива длины n
    import random
    arr = [random.randint(0, 1000) for _ in range(n)]
    
    # Замер времени
    start_time = time.time()
    bubble_sort(arr)
    end_time = time.time()
    
    return end_time - start_time

# Параметры для измерения
n_values = range(100, 10000, 100)  # Значения n от 100 до 1000 с шагом 100
times = []

# Замер времени для каждого n
for n in n_values:
    execution_time = measure_time(n)
    times.append(execution_time)
    print(f"n = {n}, Время выполнения: {execution_time:.6f} секунд")

# Построение графика
plt.plot(n_values, times, marker='o', linestyle='-', color='b')
plt.title("Время выполнения сортировки пузырьком")
plt.xlabel("Размер массива (n)")
plt.ylabel("Время выполнения (секунды)")
plt.grid(True)
plt.show()
```

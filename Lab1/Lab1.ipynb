Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер

```python
import random
import time



def generate_random_matrix(m=3, n=3, min_limit=0, max_limit=10):
    return [[random.randint(min_limit, max_limit) for _ in range(n)] for _ in range(m)]

row_sizes = [10,25, 50,75, 100,150, 200,250,300]  # Количество строк
col_sizes = [10,25, 50,75, 100,150, 200, 250, 300,350,450, 500, 600, 700,800,900, 1000, 1100,1250,1400,1500]  # Количество столбов

min_value = 0
max_value = 100
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python

#сортировка выбором

def selection_sort_row(row):
    n = len(row)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if row[j] < row[min_idx]:
                min_idx = j
        row[i], row[min_idx] = row[min_idx], row[i]

#сортировка матрицы

def sort_matrix(matrix):
    for row in matrix:
        selection_sort_row(row)

def measure_time(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix(matrix)
    return time.time() - start_time

results = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time(rows, cols)
        results.append((rows, cols, time_taken))
        print(f"Rows: {rows}, Cols: {cols}, Time: {time_taken:.6f} sec")

# Визуализация 
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# данные графика
x = [result[0] for result in results]  # Количество строк
y = [result[1] for result in results]  # Количество столбцов
z = [result[2] for result in results]  # Время 


ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Зависимость времени выполнения от размера матрицы')

plt.show()
```

```python
plt.plot(y, z, marker='o', linestyle='none', color='b')
plt.title('Зависимость времени выполнения от размера матрицы')
plt.xlabel('Размерность матрицы')
plt.ylabel('Время выполнения (Сек)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Сортировка вставкой
def insertion_sort_row(row):
    n = len(row)
    for i in range(1, n):
        key = row[i]
        j = i - 1
        while j >= 0 and row[j] > key:
            row[j + 1] = row[j]
            j -= 1
        row[j + 1] = key

def sort_matrix_insertion(matrix):
    for row in matrix:
        insertion_sort_row(row)


# Измерение времени 
def measure_time_insertion(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix_insertion(matrix)
    return time.time() - start_time


# Замер времени для всех комбинаций
results_insertion = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time_insertion(rows, cols)
        results_insertion.append((rows, cols, time_taken))
        print(f"Rows: {rows}, Cols: {cols}, Time (insertion): {time_taken:.6f} sec")

# Визуализация результатов
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = [res[0] for res in results_insertion]
y = [res[1] for res in results_insertion]
z = [res[2] for res in results_insertion]

ax.scatter(x, y, z, c='blue', marker='^')
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Сортировка вставкой: Зависимость времени от размера матрицы')

plt.show()
```

```python
# Сортировка пузырьком для одной строки
def bubble_sort_row(row):
    n = len(row)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if row[j] > row[j + 1]:
                row[j], row[j + 1] = row[j + 1], row[j]
                swapped = True
        if not swapped:
            break  # algorithm ends if the list is sorted

# Сортировка всей матрицы
def sort_matrix_bubble(matrix):
    for row in matrix:
        bubble_sort_row(row)



def measure_time_bubble(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix_bubble(matrix)
    return time.time() - start_time


results_bubble = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time_bubble(rows, cols)
        results_bubble.append((rows, cols, time_taken))
        print(f"Rows: {rows}, Cols: {cols}, Time (bubble): {time_taken:.6f} sec")


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = [res[0] for res in results_bubble]
y = [res[1] for res in results_bubble]
z = [res[2] for res in results_bubble]

ax.scatter(x, y, z, c='green', marker='s', label='Bubble Sort')
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Сортировка пузырьком: Зависимость времени от размера матрицы')

plt.show()
```

```python
# Сортировка Шелла.


def shell_sort_row(row):
    n = len(row)
    gap = n // 2  # Начальный шаг

    while gap > 0:
        for i in range(gap, n):
            temp = row[i]
            j = i
            while j >= gap and row[j - gap] > temp:
                row[j] = row[j - gap]
                j -= gap
            row[j] = temp
        gap //= 2  # Уменьшаем шаг

def sort_matrix_shell(matrix):
    for row in matrix:
        shell_sort_row(row)


def measure_time_shell(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix_shell(matrix)
    return time.time() - start_time



results_shell = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time_shell(rows, cols)
        results_shell.append((rows, cols, time_taken))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = [res[0] for res in results_shell]
y = [res[1] for res in results_shell]
z = [res[2] for res in results_shell]

ax.scatter(x, y, z, c='orange', marker='*', s=100, label='Shell Sort')
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Сортировка Шелла: Зависимость времени от размера матрицы')

plt.show()
```

```python
# Итеративная быстрая сортировка 
def quick_sort_row(row):
    stack = [(0, len(row) - 1)]  # Стек для хранения границ подмассивов
    
    while stack:
        low, high = stack.pop()
        if low >= high:
            continue
        
        # pivot selection
        mid = (low + high) // 2
        pivot = sorted([row[low], row[mid], row[high]])[1]
        
        # pivot split
        i = low - 1
        j = high + 1
        while True:
            i += 1
            while row[i] < pivot:
                i += 1
            j -= 1
            while row[j] > pivot:
                j -= 1
            if i >= j:
                break
            row[i], row[j] = row[j], row[i]
        
        # subarrays to stack
        stack.append((low, j))
        stack.append((j + 1, high))

def sort_matrix_quick(matrix):
    for row in matrix:
        quick_sort_row(row)



def measure_time_quick(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix_quick(matrix)
    return time.time() - start_time



results_quick = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time_quick(rows, cols)
        results_quick.append((rows, cols, time_taken))
        print(f"Rows: {rows}, Cols: {cols}, Time (Quick): {time_taken:.6f} sec")

# Визуализация результатов
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = [res[0] for res in results_quick]
y = [res[1] for res in results_quick]
z = [res[2] for res in results_quick]

ax.scatter(x, y, z, c='magenta', marker='X', s=100, label='Quick Sort')
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Быстрая сортировка: Зависимость времени от размера матрицы')

plt.show()
    
```

 O(n log n)

```python

class TournamentTree:
    def __init__(self, data):
        self.n = len(data)
        self.size = 1
        while self.size < self.n:
            self.size <<= 1
        self.tree = [float('inf')] * (2 * self.size)
        for i in range(self.n):
            self.tree[self.size + i] = data[i]
        for i in range(self.size - 1, 0, -1):
            self.tree[i] = min(self.tree[2 * i], self.tree[2 * i + 1])

    def get_min(self):
        return self.tree[1]

    def replace_min(self, new_val):
        idx = 1
        while idx < self.size:
            left = 2 * idx
            right = 2 * idx + 1
            if self.tree[left] == self.tree[idx]:
                idx = left
            else:
                idx = right
        self.tree[idx] = new_val
        while idx > 1:
            idx >>= 1
            new_min = min(self.tree[2 * idx], self.tree[2 * idx + 1])
            if self.tree[idx] == new_min:
                break
            self.tree[idx] = new_min

# Турнирная сортировка для одной строки
def tournament_sort_row(row):
    n = len(row)
    if n == 0:
        return
    
    tree = TournamentTree(row.copy())
    sorted_arr = []
    
    for _ in range(n):
        min_val = tree.get_min()
        sorted_arr.append(min_val)
        tree.replace_min(float('inf'))
    
    for i in range(n):
        row[i] = sorted_arr[i]

def sort_matrix_tournament(matrix):
    for row in matrix:
        tournament_sort_row(row)


def measure_time_tournament(rows, cols):
    matrix = generate_random_matrix(rows, cols)
    start_time = time.time()
    sort_matrix_tournament(matrix)
    return time.time() - start_time



results_tournament = []
for rows in row_sizes:
    for cols in col_sizes:
        time_taken = measure_time_tournament(rows, cols)
        results_tournament.append((rows, cols, time_taken))


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = [res[0] for res in results_tournament]
y = [res[1] for res in results_tournament]
z = [res[2] for res in results_tournament]

ax.scatter(x, y, z, c='brown', marker='D', s=100)
ax.set_xlabel('Количество строк')
ax.set_ylabel('Количество столбцов')
ax.set_zlabel('Время выполнения (сек)')
ax.set_title('Турнирная сортировка: Зависимость времени от размера матрицы')

plt.show()
```
